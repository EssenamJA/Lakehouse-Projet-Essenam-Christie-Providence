```python
import requests
import psycopg2

S3_URL = "http://51.77.215.42:9010"
AUTH = ("studentSDV", "coucou44")

PG_CONN = {
    "host": "51.77.215.42",
    "port": 6543,
    "dbname": "wintershop_student",
    "user": "wintershop",
    "password": "wintershop"
}

def ingest():
    conn = psycopg2.connect(**PG_CONN)
    cur = conn.cursor()

    # récupérer la liste des fichiers
    r = requests.get(S3_URL, auth=AUTH)
    files = r.text.splitlines()

    for filename in files:
        file_content = requests.get(f"{S3_URL}/{filename}", auth=AUTH).text

        for line in file_content.splitlines():
            cur.execute(
                "INSERT INTO wintershop_prod.bronze (log) VALUES (%s)",
                (line,)
            )

    conn.commit()
    cur.close()
    conn.close()

if __name__ == "__main__":
    ingest()
```
